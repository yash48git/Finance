Let's start by importing the `yfinance` library and creating a Ticker object for a stock.

```python
import yfinance as yf

# Create a Ticker object for Apple (AAPL)
# You can replace 'AAPL' with any other stock ticker symbol
stock_ticker = "AAPL"
apple = yf.Ticker(stock_ticker)

print(f"Created Ticker object for {stock_ticker}")
```

Now, let's get some basic information about the company.

```python
# Get company information
company_info = apple.info

print("\nCompany Information:")
# Display some key information from the info dictionary
if company_info:
    print(f"Company Name: {company_info.get('longName', 'N/A')}")
    print(f"Sector: {company_info.get('sector', 'N/A')}")
    print(f"Industry: {company_info.get('industry', 'N/A')}")
    print(f"Country: {company_info.get('country', 'N/A')}")
    print(f"Website: {company_info.get('website', 'N/A')}")
else:
    print("Company information not available.")
```

We can also fetch historical market data for the stock.

```python
# Get historical market data (e.g., for the last 6 months)
# The period parameter can be '1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max'
historical_data = apple.history(period="6mo")

print("\nHistorical Data (last 6 months):")
display(historical_data.head())
```

You can also specify a start and end date for historical data.

```python
# Get historical data for a specific date range
start_date = "2023-01-01"
end_date = "2023-12-31"
historical_data_range = apple.history(start=start_date, end=end_date)

print(f"\nHistorical Data ({start_date} to {end_date}):")
display(historical_data_range.head())
```

Let's retrieve the stock's current recommendation trends.

```python
# Get recommendation trends
recommendations = apple.recommendations

print("\nRecommendation Trends:")
if recommendations is not None and not recommendations.empty:
    display(recommendations)
else:
    print("Recommendation trends not available.")
```

Finally, let's get the major and institutional holders of the stock.

```python
# Get major holders
major_holders = apple.major_holders

print("\nMajor Holders:")
if major_holders is not None and not major_holders.empty:
    display(major_holders)
else:
    print("Major holders data not available.")

# Get institutional holders
institutional_holders = apple.institutional_holders

print("\nInstitutional Holders:")
if institutional_holders is not None and not institutional_holders.empty:
    display(institutional_holders)
else:
    print("Institutional holders data not available.")
```

```python
import yfinance as yf

# Replace 'AAPL' with the stock ticker you want to track
stock_ticker = "GOOGL" # Using GOOGL for Alphabet Inc. (Google)
stock = yf.Ticker(stock_ticker)

# Get and display key financial details
print(f"\nFinancial details for {stock_ticker}:")

# Get income statement
income_statement = stock.income_stmt
if income_statement is not None and not income_statement.empty:
    print("\nIncome Statement:")
    display(income_statement)
else:
    print("\nIncome Statement not available.")

# Get balance sheet
balance_sheet = stock.balance_sheet
if balance_sheet is not None and not balance_sheet.empty:
    print("\nBalance Sheet:")
    display(balance_sheet)
else:
    print("\nBalance Sheet not available.")

# Get cash flow statement
cash_flow = stock.cashflow
if cash_flow is not None and not cash_flow.empty:
    print("\nCash Flow Statement:")
    display(cash_flow)
else:
    print("\nCash Flow Statement not available.")

# Get major holders
major_holders = stock.major_holders
if major_holders is not None and not major_holders.empty:
    print("\nMajor Holders:")
    display(major_holders)
else:
    print("\nMajor Holders not available.")

# Get institutional holders
institutional_holders = stock.institutional_holders
if institutional_holders is not None and not institutional_holders.empty:
    print("\nInstitutional Holders:")
    display(institutional_holders)
else:
    print("\nInstitutional Holders not available.")
```
